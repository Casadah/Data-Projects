```python
#Import Libraries
from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
```

```python
#Connect to website
URL = 'https://www.jumia.com.ng/polo-ralph-lauren-mens-graphic-crew-t-shirt-228161365.html'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0 Safari/537.36"}

page = requests.get(URL, headers=headers)
soup1 = BeautifulSoup(page.content, "html.parser")
soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
```

```python
# Find and extract text from website
title = soup2.find('h1', class_ = "-fs20 -pts -pbxs").get_text().strip()
price = soup2.find('span', class_ = "-b -ubpt -tal -fs24 -prxs").get_text().strip()[2:]
ratings = soup2.find('div', class_ = "stars _m _al").get_text().strip()
weight = soup2.find_all('li', class_ = "-pvxs")[2].get_text().split(':')[-1].strip()
color = soup2.find_all('li', class_ = "-pvxs")[3].get_text().split(':')[-1].strip()
unit_left = soup2.find('p', class_ = "-df -i-ctr -fs12 -pbs -rd5").get_text().strip()

print(title)
print(price)
print(ratings)
print(weight)
print(color)
print(unit_left)
```

```python
# Adding date to data
today = datetime.date.today()

print(today)
```

```python
# Saving extracted to CSV
import csv

header = ['Product', 'Price (₦)', 'Weight (kg)', 'Color', 'Rating', 'Unit Left', 'Date']
data = [title, price, weight, color, ratings, unit_left, today]

with open('OnlineStoreWebScrapping.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```
